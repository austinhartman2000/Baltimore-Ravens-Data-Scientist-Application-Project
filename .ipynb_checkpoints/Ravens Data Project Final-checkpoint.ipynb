```python
#Austin Hartman 5/28/2026

#The goal of this project is to assess the ability of a corner to prevent targets while in man coverage on a WR
#I'm not analyzing anything that happens after the ball is thrown, I am purely evaluating lockdown/coverage ability to prevent targets

import pandas as pd
from functools import reduce
import numpy as np
import gpboost as gpb
from sklearn.model_selection import train_test_split
from sklearn.calibration import calibration_curve
import matplotlib.pyplot as plt
from sklearn.metrics import (
    roc_auc_score, log_loss, brier_score_loss,
    classification_report, confusion_matrix, 
    ConfusionMatrixDisplay, RocCurveDisplay,
    PrecisionRecallDisplay, average_precision_score
)
from sklearn.metrics import f1_score, precision_score, recall_score

#Run this cell first for all dependencies

#For quick-running the code with the fully built data object, skip to to the data pickle load later on
#To build the data from ground up, start at the beginning
```

```python
#Load all datasets
games = pd.read_csv("data/games.csv")
plays = pd.read_csv("data/plays.csv")
players = pd.read_csv("data/players.csv")
player_play = pd.read_csv("data/player_play.csv")
tracking_data = pd.concat(
    [pd.read_csv(f"data/tracking_week_{week}.csv") for week in range(1, 10)],
    ignore_index=True
)
    
```

```python
#Isolate only plays with CBs playing Man coverage
corners = players[players["position"] == "CB"]
player_play_man = player_play[(player_play["nflId"].isin(corners["nflId"])) & (player_play["pff_defensiveCoverageAssignment"] == "MAN")]
player_play_man = player_play_man.merge(players[["nflId", "displayName"]], on = "nflId", how = "left")
player_play_man
```

```python
#Filter, only plays classified as man coverage, removes a couple weird hybrid plays with man-coverage individuals but zone overall play
man_coverage = plays[(plays["pff_manZone"] == "Man")]
filtered_player_plays = man_coverage.merge(
    player_play_man, 
    on=["gameId", "playId"], 
    how="inner"
)
filtered_player_plays
```

```python
#Receiver play and player info
receiver_play = player_play.add_suffix("_receiver").rename(
    columns = {"gameId_receiver": "gameId", "playId_receiver": "playId"})
receiver_player = players.add_suffix("_receiver")
    
#Merge WR info based on Man coverage primary matchup assignment
df_wr1 = filtered_player_plays.merge(
    receiver_play,
    left_on=["gameId", "playId", "pff_primaryDefensiveCoverageMatchupNflId"],
    right_on=["gameId", "playId", "nflId_receiver"],
    how="left"
)
df_wr = df_wr1.merge(
    receiver_player,
    on = ["nflId_receiver"],
    how = "left"
)

#Filter out RB and TE routes, different coverage styles than WRs
df_wr = df_wr[df_wr["position_receiver"] == "WR"]
df_wr
```

```python
#WR tracking data for target depth
tracking_wr = tracking_data.add_suffix("_receiver").rename(
    columns={"gameId_receiver": "gameId", "playId_receiver": "playId", "frameId_receiver": "frameId"}
).drop("displayName_receiver", axis = 1)

filtered_tracking_data = df_wr.merge(
    tracking_wr, 
    left_on=["gameId", "playId", "pff_primaryDefensiveCoverageMatchupNflId"],
    right_on=["gameId", "playId", "nflId_receiver"],
    how="inner"
)

#Simplify naming conventions
filtered_tracking_data = filtered_tracking_data.rename(
    columns={"pff_primaryDefensiveCoverageMatchupNflId": "nflId_receiver"}
)

#Find the frames with forward passes, the moment of the target
pass_frames = (
    filtered_tracking_data[filtered_tracking_data["event_receiver"] == "pass_forward"]
    .groupby(["gameId", "playId"])["frameId"]
    .min()
    .rename("pass_forward_frame")
)

#If there is no pass thrown, when did the play end?
other_frames = (
    filtered_tracking_data[filtered_tracking_data["event_receiver"].isin(["qb_sack", "run", "out_of_bounds", "qb_strip_sack", "safety"])]
    .groupby(["gameId", "playId"])["frameId"]
    .min()
    .rename("other_frame")
)

dfs = [filtered_tracking_data, pass_frames, other_frames]
filtered_tracking_data = reduce(lambda left, right: pd.merge(left, right, on=["gameId", "playId"], how='outer'), dfs)

#Determine the last frame before throw or play end
filtered_tracking_data["target_frame"] = filtered_tracking_data["pass_forward_frame"].fillna(filtered_tracking_data["other_frame"])
filtered_tracking_data
```

```python
#Pull only needed info from tracking data to reduce size
target_frame_tracking = (
    filtered_tracking_data[
        ["gameId", "playId", "nflId", "displayName", "displayName_receiver", "frameId", "x_receiver", "target_frame", "playDirection_receiver", "receivingYards_receiver", "absoluteYardlineNumber", "nflId_receiver", "routeRan_receiver", "dropbackType", "playAction", "timeToThrow", "wasTargettedReceiver_receiver"]
    ]
)

#Filter to only frame rows when the ball is being thrown/target ocurring
summary = target_frame_tracking[
    target_frame_tracking["frameId"] == target_frame_tracking["target_frame"]
].copy()

#Compute depth of receiver at time of throw
summary["depth_at_target"] = np.where(
    summary["playDirection_receiver"] == "right",
    summary["x_receiver"] - summary["absoluteYardlineNumber"],
    summary["absoluteYardlineNumber"] - summary["x_receiver"]
)

summary
```

```python
summary = summary.sort_values(["nflId_receiver", "gameId", "playId"]).copy()

#Robustness check for basic variables
summary["target"] = summary["wasTargettedReceiver_receiver"].astype(int)
summary["yards"] = summary["receivingYards_receiver"].fillna(0)

#Depth of target (only when targetted) for aDot calculations
summary["wr_dot"] = summary["target"] * summary["depth_at_target"]

#Gather cumulative stats up until the current play
#Shift all by 1 to prevent data leakage - rows include their own information otherwise
summary["wr_targets"] = (
    summary.groupby("nflId_receiver")["target"]
    .cumsum().shift(1)  # exclude current play
)
summary["wr_yards"] = (
    summary.groupby("nflId_receiver")["yards"]
    .cumsum().shift(1)
)
summary["wr_routes"] = (
    summary.groupby("nflId_receiver")
    .cumcount()  # cumcount already excludes current row so no shift needed
)
summary["wr_sum_dot"] = (
    summary.groupby("nflId_receiver")["wr_dot"]
    .cumsum().shift(1)
)

#Fill NaN with 0 for first play due to shift
summary["wr_targets"] = summary["wr_targets"].fillna(0)
summary["wr_yards"] = summary["wr_yards"].fillna(0)
summary["wr_sum_dot"] = summary["wr_sum_dot"].fillna(0)

#League averages to shrink towards for small samples
league_tprr = summary["target"].mean()
league_yprr = summary["yards"].mean()

league_adot = (
    summary.loc[summary["target"] == 1, "depth_at_target"].mean()
)

alpha = 20
n = summary["wr_routes"]

#Shrunk Targets per Route Run
summary["wr_shrunk_tprr"] = (
    (summary["wr_targets"] + alpha * league_tprr) /
    (n + alpha)
)

#Shrunk Yards per Route Run
summary["wr_shrunk_yprr"] = (
    (summary["wr_yards"] + alpha * league_yprr) /
    (n + alpha)
)

#Shrunk aDot
summary["wr_shrunk_adot"] = (
    (summary["wr_sum_dot"] + alpha * league_adot) /
    (summary["wr_targets"] + alpha)
)

summary
```

```python
#Save final dataset for model building
summary.to_pickle("summary.pkl")
```

```python
# Load to reduce wait times, don't need to recreate data every time
summary = pd.read_pickle("summary.pkl")
```

```python
#Count total routes to set minimum thresholds
cb_route_counts = (
    summary.groupby(["displayName", "nflId"])
    .size()
    .reset_index(name="n_routes")
)

print(cb_route_counts.sort_values(by = ["n_routes"]))
```

```python
#Define features
features = [
    "routeRan_receiver", "dropbackType", "timeToThrow",
    "depth_at_target", "playAction", "wr_shrunk_tprr", "wr_shrunk_yprr", "wr_shrunk_adot"
]

target = "wasTargettedReceiver_receiver"

#Drop NAs
model_df = summary[features + [target, "nflId", "nflId_receiver"]].dropna()

model_df["routeRan_receiver"] = (
    model_df["routeRan_receiver"].astype("category")
)

model_df["dropbackType"] = (
    model_df["dropbackType"].astype("category")
)

X = model_df[features]
y = model_df[target]
groups = model_df[["nflId", "nflId_receiver"]].values

X_train, X_test_temp, y_train, y_test_temp, groups_train, groups_test_temp = train_test_split(
    X, y, groups, test_size=0.2, random_state=1
)

#Small validation set just for setting classification cutoff
X_test, X_val, y_test, y_val, groups_test, groups_val = train_test_split(
    X_test_temp, y_test_temp, groups_test_temp, test_size=0.25, random_state=1
)

#Define random effects model — one random effect per CB
gp_model = gpb.GPModel(group_data=groups_train, likelihood="bernoulli_logit")

#Build dataset
data_train = gpb.Dataset(X_train, y_train, categorical_feature=[
        "routeRan_receiver",
        "dropbackType"
    ])

#Parameters for model
params = {
    "objective": "binary",
    "learning_rate": 0.05,
    "max_depth": 4,
    "num_leaves": 31,
    "min_data_in_leaf": 5,
    "verbose": -1
}

model = gpb.train(
    params=params,
    train_set=data_train,
    gp_model=gp_model,
    num_boost_round=100
)
```

```python
#Calculate feature importance
importance = pd.DataFrame({
    "feature": features,
    "importance": model.feature_importance(importance_type="gain")
}).sort_values("importance", ascending=False)

print(importance)
```

```python
unique_cbs = pd.DataFrame({
    "nflId": model_df["nflId"].unique()
})

#Hold receiver constant
reference_receiver = model_df["nflId_receiver"].iloc[0]
cb_pred_groups = pd.DataFrame({
    "nflId": unique_cbs["nflId"],
    "nflId_receiver": reference_receiver
})

#Predict random effects
cb_preds = gp_model.predict(
    group_data_pred=cb_pred_groups,
    predict_var=False
)

#Build dataframe
cb_effect_df = pd.DataFrame({
    "nflId": unique_cbs["nflId"],
    "target_prob": cb_preds["mu"],
})
cb_effect_df["log_odds"] = np.log(
    cb_effect_df["target_prob"] / (1 - cb_effect_df["target_prob"])
)
baseline = cb_effect_df["log_odds"].mean()
cb_effect_df["log_odds_vs_avg"] = cb_effect_df["log_odds"] - baseline

#Z-score scaled 0-100 skill score (higher = better CB)
mean = cb_effect_df["log_odds"].mean()
std = cb_effect_df["log_odds"].std()
cb_effect_df["skill_score"] = (
    50 - ((cb_effect_df["log_odds"] - mean) / std) * 15
).clip(0, 100)

#Percentile rank (100 = best CB)
cb_effect_df["percentile"] = (
    cb_effect_df["log_odds"]
    .rank(pct=True, ascending=False) * 100
).round(1)

#Add names and sort
cb_effect_df = cb_effect_df.merge(
    summary[["nflId", "displayName"]].drop_duplicates(),
    on="nflId"
).sort_values("skill_score", ascending=False)

print(cb_effect_df[["displayName", "target_prob", "log_odds_vs_avg", "skill_score", "percentile"]])
```

```python
#Subset on players with at least 30 routes defended and save and display final output
cb_effect_routes_df = cb_effect_df.merge(cb_route_counts.drop("displayName", axis = 1), on = ["nflId"])
cb_effect_routes_df = cb_effect_routes_df[cb_effect_routes_df["n_routes"] >= 30]
cb_effect_routes_df.to_csv("Output.csv")

cb_effect_routes_df[0:20]
```

```python
y_pred_prob = model.predict(
    data=X_test, 
    group_data_pred=groups_test
)["response_mean"]

prob_true, prob_pred = calibration_curve(y_test, y_pred_prob, n_bins=10)

# Get sample counts per bin
bin_edges = np.linspace(0, 1, 11)
bin_counts, _ = np.histogram(y_pred_prob, bins=bin_edges)
bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2

# Mask for bars — applied to the 10-element bin arrays
min_samples = 20
bar_mask = bin_counts >= min_samples

# Mask for calibration line — match prob_pred points to bins
cal_bin_idx = np.searchsorted(bin_edges[1:], prob_pred, side="left")
cal_bin_idx = np.clip(cal_bin_idx, 0, len(bin_counts) - 1)
cal_mask = bin_counts[cal_bin_idx] >= min_samples

fig, ax1 = plt.subplots(figsize=(8, 6))

ax2 = ax1.twinx()
ax2.bar(
    bin_centers[bar_mask],
    bin_counts[bar_mask],
    width=0.08,
    alpha=0.2,
    color="tab:orange",
    label="Sample size"
)
ax2.set_ylabel("Sample Count", color="gray")
ax2.tick_params(axis="y", labelcolor="gray")

ax1.plot(prob_pred[cal_mask], prob_true[cal_mask], marker="o", label="GPBoost", color="steelblue", zorder=5)
ax1.plot([0, 1], [0, 1], linestyle="--", color="gray", label="Perfect calibration", zorder=4)
ax1.fill_between(prob_pred[cal_mask], prob_true[cal_mask], prob_pred[cal_mask], alpha=0.1, color="steelblue")
ax1.set_xlabel("Mean Predicted Probability")
ax1.set_ylabel("Fraction of Positives (Actual Target Rate)")
ax1.set_title("Calibration Plot — Target Probability")
ax1.set_xlim(0, 1)
ax1.set_ylim(0, 1)
ax1.legend(loc="upper left")
ax1.grid(alpha=0.3)

plt.tight_layout()
plt.savefig("Calibration Plot.png")
plt.show()
```

```python
#Use validation set for setting the decision cutoff

thresholds = np.arange(0.1, 0.9, 0.01)

y_pred_prob_val = model.predict(
    data=X_val, 
    group_data_pred=groups_val
)["response_mean"]

results = pd.DataFrame([
    {
        "threshold": t,
        "f1": f1_score(y_val, (y_pred_prob_val >= t).astype(int), zero_division=0),
        "precision": precision_score(y_val, (y_pred_prob_val >= t).astype(int), zero_division=0),
        "recall": recall_score(y_val, (y_pred_prob_val >= t).astype(int), zero_division=0)
    }
    for t in thresholds
])

best_threshold = results.loc[results["f1"].idxmax(), "threshold"]
print(f"Optimal threshold: {best_threshold:.2f}")
print(results.loc[results["f1"].idxmax()])

# Plot all three metrics across thresholds
plt.figure(figsize=(10, 5))
plt.plot(results["threshold"], results["f1"], label="F1", color="steelblue")
plt.plot(results["threshold"], results["precision"], label="Precision", color="darkorange")
plt.plot(results["threshold"], results["recall"], label="Recall", color="green")
plt.axvline(best_threshold, linestyle="--", color="gray", label=f"Optimal ({best_threshold:.2f})")
plt.xlabel("Threshold")
plt.ylabel("Score")
plt.title("Metrics vs Classification Threshold")
plt.legend()
plt.grid(alpha=0.3)
plt.savefig("classification Threshold.png")
plt.show()

# Apply optimal threshold
y_pred_optimal = (y_pred_prob_val >= best_threshold).astype(int)
print(classification_report(y_val, y_pred_optimal, target_names=["Not Targeted", "Targeted"]))
```

```python
#Accuracy Metrics
print(f"ROC-AUC:        {roc_auc_score(y_test, y_pred_prob):.4f}")
print(f"Log Loss:       {log_loss(y_test, y_pred_prob):.4f}")
print(f"Brier Score:    {brier_score_loss(y_test, y_pred_prob):.4f}")
print(f"Avg Precision:  {average_precision_score(y_test, y_pred_prob):.4f}")
print("\n", classification_report(y_test, y_pred_optimal, target_names=["Not Targeted", "Targeted"]))

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

#ROC Curve
RocCurveDisplay.from_predictions(y_test, y_pred_prob, ax=axes[0, 0], name="GPBoost")
axes[0, 0].plot([0, 1], [0, 1], linestyle="--", color="gray")
axes[0, 0].set_title("ROC Curve")

#Precision-Recall Curve
PrecisionRecallDisplay.from_predictions(y_test, y_pred_prob, ax=axes[0, 1], name="GPBoost")
baseline = y_test.mean()
axes[0, 1].axhline(baseline, linestyle="--", color="gray", label=f"Baseline ({baseline:.2f})")
axes[0, 1].set_title("Precision-Recall Curve")
axes[0, 1].legend()

#Confusion Matrix
ConfusionMatrixDisplay(
    confusion_matrix(y_test, y_pred_optimal),
    display_labels=["Not Targeted", "Targeted"]
).plot(ax=axes[1, 0], colorbar=False)
axes[1, 0].set_title("Confusion Matrix")

#Predicted Probability Distribution by Class
axes[1, 1].hist(
    y_pred_prob[y_test == 0], bins=30,
    alpha=0.6, color="steelblue", label="Not Targeted", density=True
)
axes[1, 1].hist(
    y_pred_prob[y_test == 1], bins=30,
    alpha=0.6, color="darkorange", label="Targeted", density=True
)
axes[1, 1].axvline(0.5, linestyle="--", color="gray")
axes[1, 1].set_xlabel("Predicted Probability")
axes[1, 1].set_ylabel("Density")
axes[1, 1].set_title("Predicted Probability Distribution by Class")
axes[1, 1].legend()

plt.tight_layout()
plt.savefig("Results graphs.png")
plt.show()
```
